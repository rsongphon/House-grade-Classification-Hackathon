```python
import os
```

```python
class1_dark = os.listdir('/content/class_1_CSV/class 1/class1_dark')
class1_flip = os.listdir('/content/class_1_CSV/class 1/class1_flip')
class1_fog = os.listdir('/content/class_1_CSV/class 1/class1_fog')
class1_rain = os.listdir('/content/class_1_CSV/class 1/class1_rain')
class1_speed = os.listdir('/content/class_1_CSV/class 1/class1_speed')
class1_sun = os.listdir('/content/class_1_CSV/class 1/class1_sun')
```

```python
class1_all = class1_dark+class1_flip+class1_fog+class1_rain+class1_speed+class1_sun
```

```python
len(class1_all)
```

```python
class1_label = [int(1) for img in class1_all]
```

```python
import pandas as pd
```

```python
train_df = pd.read_csv('/content/house-grade/train.csv')
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
class1_dict = {
    'image_name' : class1_all , 
    'class' : class1_label
}
```

```python
class1_df = pd.DataFrame(class1_dict)
```

```python
new_train_df = pd.concat([train_df,class1_df], axis=0)
```

```python
new_train_df.info()
```

```python
new_train_df.tail()
```
